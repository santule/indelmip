```python
from pysam import FastaFile,FastxFile
from ete3 import Tree
import pandas as pd
```

```python
# read the mip indel ancestor file
folder_location         = '/Users/sanjanatule/Documents/uq/Projects/MIPIndel/data/'

## Sample tree 1
tree_name               = 'CYP2U_165'
nwk_file_path           = folder_location + tree_name + '/CYP2U_165.nwk'
mip_indel_file          = folder_location + tree_name + '/mip_ancestor_indel.fasta'
```

```python
# get the indel patterns and give them unique number
sequence_pattern_dict = {}
sequence_pattern_index = {}
sequence_name_pattern_index = {}
pattern = 1
with FastxFile(mip_indel_file) as fh:
    for entry in fh:
        # add start and end string to the sequence
        seq_name = entry.name
        indel_sequence = entry.sequence[1:-1] # remove start and end node
        if sequence_pattern_dict.get(indel_sequence):
            sequence_name_pattern_index[seq_name] = sequence_pattern_dict[indel_sequence]
        else:
            sequence_pattern_dict[indel_sequence] = pattern
            sequence_pattern_index[pattern]       = indel_sequence
            sequence_name_pattern_index[seq_name] = pattern
            pattern = pattern + 1 
```

```python
# get the sequence levels in the tree
tree_file = open(nwk_file_path,"r")
my_tree = tree_file.read() + ";"
tree = Tree(my_tree, format=1)

sequence_tree_level = {}

edge = 0
for n in tree.traverse():
    if not n.is_leaf():
        n.name = "NODE_%d" %edge
        edge += 1
level = 0                
for n in tree.traverse():
    if n.up is not None: # root node
        n.add_features(level = n.up.level + 1)
        sequence_tree_level[n.name] = n.up.level + 1
    else:
        n.add_features(level = level)
        sequence_tree_level[n.name] = level
```

```python
# wrangle data for visualisation
indel_visual_df1 = pd.DataFrame(sequence_tree_level.items(),columns = ['sequence','tree_level'])
indel_visual_df2 = pd.DataFrame(sequence_name_pattern_index.items(),columns = ['sequence','pattern_index'])
indel_visual_df = pd.merge(indel_visual_df2, indel_visual_df1, how= "inner" , on="sequence")
indel_visual_df = indel_visual_df.drop(columns='sequence')
indel_visual_df['dummy_value'] = 1
indel_visual_df =indel_visual_df.drop_duplicates()
indel_visual_df = indel_visual_df.pivot(index='pattern_index', columns= 'tree_level',values = 'dummy_value')
indel_visual_df = indel_visual_df.fillna(0)
indel_visual_df = indel_visual_df.astype(int)
```

```python
# heatmap
import seaborn as sns
colors = ["#F5F3EE", "#B173AE"]
sns.set(rc = {'figure.figsize':(15,10)})
ax = sns.heatmap(indel_visual_df,cmap = colors,cbar=False)
ax.set(xlabel='Tree Level' , ylabel='Indel Pattern')
```
