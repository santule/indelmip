```python
import pandas as pd
import seaborn as sns
```

# SCORE

```python
score_data = pd.read_csv('/content/drive/MyDrive/UQ/projects/MIPIndel/score.csv')
score_data = score_data.melt(id_vars=['Protein Family'], value_vars=['MIPIndel', 'BEP','SICP','PSP'])
score_data.head(10)
```

```python
g = sns.catplot(data = score_data, kind='bar', x="Protein Family", y="value", hue="variable", palette="dark", alpha=.6, height=10, aspect=11.7/8.27)
g.despine(left=True)
g.set_axis_labels("", "Indel Score")
g.legend.set_title("")
```

# NEW/OLD PATTERNS

```python
new_old_p_data = pd.read_csv('/content/drive/MyDrive/UQ/projects/MIPIndel/old_new_pattern.csv')
new_old_p_data['New%'] = new_old_p_data['New'] / (new_old_p_data['Old'] + new_old_p_data['New'])
new_old_p_data.head(10)
```

```python

g = sns.catplot(data = new_old_p_data, kind='bar', x="Protein Family", y="New%", hue="Method", palette="dark", alpha=.6, height=10, aspect=11.7/8.27)
g.despine(left=True)
g.set_axis_labels("", "New Ancestor Indel Patterns")
g.legend.set_title("")
```
